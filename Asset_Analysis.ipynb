```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def assent_value_calc(e, sigma, N):

    sharpe_ratio = 0.2

    # Generate data
    x = np.arange(N+1)
    expected_growth = e * 100 * x 


    upper_bound = 100 * np.exp((e - 0.5*sigma**2)*x + 3*sigma*np.sqrt(x)) - 100
    lower_bound = 100 * np.exp((e - 0.5*sigma**2)*x - 3*sigma*np.sqrt(x)) - 100

    # # Generate a mock 'actual' curve
    np.random.seed(40)  # for reproducibility
    actual_growth = 100 * np.cumprod(1 + np.random.normal(e, sigma, N+1)) - 100

    # # Create the plot
    plt.figure(figsize=(12, 8))
    plt.plot(x, expected_growth, label='Expected Growth Curve', color='blue')
    plt.plot(x, actual_growth, label='Actual Asset Curve', color='black', linewidth=2)
    plt.plot(x, upper_bound, '--', color='red', alpha=0.5, label='+3σ Curve')
    plt.plot(x, lower_bound, '--', color='red', alpha=0.5, label='-3σ Curve')

    # # Set labels and title
    plt.xlabel('Number of Trades')
    plt.ylabel('Asset Ratio [%]')
    plt.title(f'Expected Return e={e*100}%, Standard Deviation σ={sigma*100}% (Sharpe Ratio={sharpe_ratio})\n'
              f'Number of Trades per Unit Period N={N}')

    # Set y-axis limits and grid
    plt.ylim(-20, 100)
    plt.grid(True)

    # Add legend
    plt.legend()

    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
e = 0.002
sigma = 0.01
N = 300
assent_value_calc(e, sigma, N)
```

## A formula to measure if the strategy satisfied the trading goal.
Ｅ≧target return，D≦tolerance
A = E/D
D = 9/4×σ^2／e
E = e * N

4/9 * N * (e/σ)^2 >= A

```python
e = 0.002 * 100
sigma = 0.01 * 100
N = 300

target = e * N
tolerance = 10

calc = 4/9 * N * (e / sigma) ** 2
A = target / tolerance

print(calc)
print(A)
```
